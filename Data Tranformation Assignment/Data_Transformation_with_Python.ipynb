```python
IQR = dataset['Employer Health Insurance Coverage (2015)'].quantile(0.75) - dataset['Employer Health Insurance Coverage (2015)'].quantile(0.25)
IQR
```

```python
Higher_outlier = dataset['Employer Health Insurance Coverage (2015)'].quantile(0.75) + (IQR * 1.5)
lower_outlier = dataset['Employer Health Insurance Coverage (2015)'].quantile(0.25) - (IQR * 1.5)
print(Higher_outlier)
print(lower_outlier)
```

```python
dataset.loc[dataset['Employer Health Insurance Coverage (2015)'] >= Higher_outlier, 'Employer Health Insurance Coverage (2015)']
```

```python
dataset.loc[dataset['Employer Health Insurance Coverage (2015)'] >= Higher_outlier, 'Employer Health Insurance Coverage (2015)'] = 9254000
```
