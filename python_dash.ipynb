```python
# scatter diagram

from plotnine import ggplot, aes, geom_point, theme_minimal, labs

# Assuming 'dataset' is your DataFrame
# Selecting variables for the scatter plot
employer_coverage_2015 = 'Employer Health Insurance Coverage (2015)'
medicare_enrollment_2016 = 'Medicare Enrollment (2016)'

# Create ggplot
scatter_plot = (
    ggplot(dataset) +
    aes(x=employer_coverage_2015, y=medicare_enrollment_2016) +
    geom_point() +
    theme_minimal() +
    labs(title="Scatter Plot of Employer Health Insurance Coverage (2015) vs. Medicare Enrollment (2016)",
         x="Employer Health Insurance Coverage (2015)",
         y="Medicare Enrollment (2016)")
)

# Show the plot
print(scatter_plot)
```
